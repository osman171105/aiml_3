```python
model1 = smf.ols("sunday~daily",data = data1).fit()
```

```python
model1.summary()
```

### Observation
- the predicted equation is y_hat = Beta_0 + Beta_1
- y_hat = 13.8356 + 1.3397X

- the probability (p-value) for intercept (beta) is 0.707 > 0.05
- there fore intercept coefficient may not be that  much significiant in prediction

- however the the p-value for "daily " (beta_1) is 0.00 < 0.05
- therefore the beta_1 coefficient is hihly significiant and is contributint to prediction

```python
model1.params
```

```python
print(f'model t-values:\n{model1.tvalues}\n-----------------\n model p-values : \n{model1.pvalues}')
```

```python
pred = model1.predict(data1['daily'])
pred
```

```python
data1['Y_hat'] = pred
data1
```

```python
data1['residuals'] = data1['sunday']-data1['Y_hat']
data1
```

```python
mse = np.mean((data1['daily']-data1['Y_hat'])**2)
rmse = np.sqrt(mse)
print('MSE  : ',mse)
print('RMSE : ',rmse)
```

```python
plt.scatter(data1['Y_hat'],data1['residuals'])
```

#### observations
- There appears to be no trend and the residuals are randomly placed around the zero eror line
- Hence the assumption of homoscedasticity is satisfied (constant variance in residuals) .
