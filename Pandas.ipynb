```python
import pandas as pd
import string
t = " One of the simplest ways to convert a dictionary to a DataFrame is by using the `pandas. DataFrame. from_dict()` method. This method takes the dictionary as input and returns a DataFrame with the dictionary keys as column names and the corresponding values as data.In this blog, we’ll guide data scientists and software engineers through the step-by-step process of creating a dictionary from two columns in a pandas DataFrame, addressing the challenges that may arise, particularly for those new to pandas or Python."
t
```

```python
text = t.lower()
text
```

```python
t = text.translate(str.maketrans(" "," ",string.punctuation))
t
```

```python
w = t.split()
w
```

```python
wc = {}
for word in w:
    wc[word] = wc.get(word,0)+1
wc
```

```python
df = pd.DataFrame(list(wc.items()),columns = ["word","count"])
print(df)
```
